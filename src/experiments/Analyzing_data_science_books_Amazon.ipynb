### Clustering book titles

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words='english', ngram_range=(1,2))
X = vectorizer.fit_transform(df["title"])
pd.DataFrame(X.toarray(), columns = vectorizer.get_feature_names())
```

```python
from sklearn.cluster import KMeans
# sum square distance
SSD = []
for k in range(2,15):
    km = KMeans(n_clusters=k, max_iter=600, n_init=10)
    km.fit(X)
    SSD.append(km.inertia_)
SSD
```

```python
plt.plot(range(2,15), SSD, 'bx-')
plt.xlabel('Number of clusters')
plt.ylabel('Sum of squared distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
model = KMeans(n_clusters=6, max_iter=600, n_init=10)
model.fit(X)
labels = model.labels_
books_cl = pd.DataFrame(list(zip(df["title"],labels)),columns=["title","cluster"])
books_cl
```

```python
from wordcloud import WordCloud
for k in range(6):
    text= books_cl[books_cl["cluster"]==k]["title"].str.cat(sep = ' ')
    mywordcloud = WordCloud(max_font_size=100,max_words=100, background_color='white').generate(text)
    plt.subplot(2,3,k+1).set_title("Cluster " + str(k))
    plt.imshow(mywordcloud,interpolation="bilinear")
plt.show()
```

```python
product_url = "https://www.amazon.com/Deep-Learning-Adaptive-Computation-Machine/dp/0262035618/"
reviews_url  = "https://www.amazon.com/Deep-Learning-Adaptive-Computation-Machine/product-reviews/0262035618/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews"
```

```python
def get_review_url(product_url):
    try:
        split_url = product_url.split('dp')
        product_number = split_url[1].split('/')[1]
        reviews_url = split_url[0] + 'product-reviews/'+ product_number + "/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews"
    except:
        reviews_url = None
    return reviews_url
```

```python
example = get_review_url(product_url)
example
```

```python
df['review_urls'] = df['complete_link'].apply(lambda x:get_review_url(x))
# remove empty reviews url
df_reviews = df.loc[~df['review_urls'].isnull()].reset_index()
df_reviews.head()
```
